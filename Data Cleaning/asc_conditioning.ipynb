### Split authors in ASC data by semicolon + space and create a single entry per author. Then check for any spaces surrounding names

```python
split_authors_df = asc_og_df.Author.str.split('; ', expand=True).stack()
full_split_df = asc_og_df.join(pd.Series(index=split_authors_df.index.droplevel(1), data=split_authors_df.values, name='Single_Author'))
```

```python
split_authors_list = split_authors_df.to_list()
split_authors_list[:10]
```

```python
full_split_df[:5]
```

### Change ordering of columns; bring 'Single_Author' column next to original 'Author' column and get rid of the 'Author' column

```python
full_split_df = full_split_df[['Article Title', 'Author', 'Single_Author', 'Journal Title', 'ISSN', 'ISBN', 
                                'Publication Date', 'Volume', 'Issue', 'First Page', 'Page Count', 'Accession Number',
                                'DOI', 'Publisher', 'Doctype', 'Subjects', 'Keywords', 'Abstract',
                                'PLink']]
```

```python
single_author_df = full_split_df.drop('Author', axis=1)
```

```python
single_author_df[:5]
```

### Create new columns for authors' first and last names; split 'Single_Author' by the comma and space

```python
single_author_df[['Author_ln', 'Author_fn']] = single_author_df['Single_Author'].str.split(', ', expand=True)
single_author_df[['Author_fn']] = single_author_df['Author_fn'].str.split(' ', expand=True)[0]
single_author_df[:5]
```

### Get rid of 'Single_Author' column, rearrange authors' first and last name columns towards first few columns of dataframe

```python
single_author_df = single_author_df[['Article Title', 'Author_fn', 'Author_ln', 'Journal Title', 'ISSN', 'ISBN', 
                                'Publication Date', 'Volume', 'Issue', 'First Page', 'Page Count', 'Accession Number',
                                'DOI', 'Publisher', 'Doctype', 'Subjects', 'Keywords', 'Abstract',
                                'PLink']]
single_author_df[:5]
```

### Combine authors' first and last names into a single column called 'Author'

```python
single_author_df['Author'] = single_author_df['Author_fn'] + ' ' + single_author_df['Author_ln']
final_df = single_author_df[['Article Title', 'Author', 'Journal Title', 'ISSN', 'ISBN', 
                                'Publication Date', 'Volume', 'Issue', 'First Page', 'Page Count', 'Accession Number',
                                'DOI', 'Publisher', 'Doctype', 'Subjects', 'Keywords', 'Abstract',
                                'PLink']]
final_df[:5]
```
